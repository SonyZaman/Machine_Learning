```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
titanic = pd.read_csv('train.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

## 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'])
```

## 2. Bar Plot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(x=titanic['Pclass'], y=titanic['Age'])
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Fare'],hue=titanic['Sex'])
```

## 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x=titanic['Survived'],y=titanic['Age'])
```

```python
sns.boxplot(x=titanic['Survived'],y=titanic['Age'],hue=titanic['Sex'])
```

## 4. Kdeplot (Numerical - Categorical)

```python
sns.kdeplot(titanic[titanic['Survived']==0]['Age'])
sns.kdeplot(titanic[titanic['Survived']==1]['Age'])
```

```python
sns.kdeplot(data=titanic, x='Age', hue='Survived')
```

## 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
(titanic.groupby('Embarked')['Survived'].mean() * 100)
```

## 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

## 7. Pairplot

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

## 8. Lineplot (Numerical - Numerical)

```python
flights.head()
```

```python
new = flights.groupby('year')['passengers'].sum().reset_index()
```

```python
sns.lineplot(x=new['year'], y=new['passengers'])
```

```python
sns.lineplot(data=new, x='year', y='passengers')
```
